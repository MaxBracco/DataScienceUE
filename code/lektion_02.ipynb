## Clustering in Scikit-learn

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
#Hier 2 gewählt, da man von Fake und nicht Fake ausgegangen ist.
kmeans_engine = KMeans(n_clusters=2)

#Einstellen und Anwenden in 2 Schritten
kmeans_engine.fit(X)
clusters = kmeans_engine.predict(X)

#Einstellen und Anwenden in 1 Schritt
# kmeans_engine.fit_predict(X)
```

```python
X["clusters"] = clusters
```

```python
X.plot()
```

```python
X.plot(
    kind='scatter',
    x='description_length',
    y='words_fullname',
    c='clusters',
    colormap='viridis',
    )
```
